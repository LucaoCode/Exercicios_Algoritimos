```python
#Exercicio 1
N1 = float(input('Digite sua nota da N1 >>>> '))
N2 = float(input('Digite sua nota da N2 >>>> '))
N3 = float(input('Digite sua nota da N3 >>>> '))
ME = float(input('Digite sua nota da ME >>>> '))
media_notas = (N1 + N2 * 2 + N3 * 3+ ME)/7
print('Sua Nota no semestre é >>> ' , round(media_notas,2))
```

```python
#Exercicio 2
import math

x1 = float(input('Digite x1 >>>> '))
x2 = float(input('Digite x2 >>>> '))
y1 = float(input('Digite y1 >>>> '))
y2 = float(input('Digite y2 >>>> '))

d = round(math.sqrt((x2-x1)**2 + (y2 - y1)**2),2)
print(d)

# tem como calcular deste modo tbm, chegará ao msm resultado = ((x2-x1)**2 + (y2 - y1)**2)**0.5
```

```python
#Exercicio 4
#percentagem = 28%
#impostos = 45%
#custoConsu = CF/porc e inpostos

#Entrada
custoDeFabrica = float(input("Digite o custo de fabrica >>> "))

#Processo
totalDistribuidor = custoDeFabrica * 0.28
totalImpostos = custoDeFabrica * 0.45
total = totalDistribuidor + totalImpostos + custoDeFabrica


#Saida
print("Custo total da fabrica >>> " , total)

```
