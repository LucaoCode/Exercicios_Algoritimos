```python
lista2 = [["jonas",2300,2,10,1],["Danielle",2600,1,5,0],["Cleiton",2500,0,7,1]]
novaLista = []
dist = []

nome = input("Digite seu nome >>> ")
novaLista.append(nome)

salario = float(input("Digite seu salario >>> "))
novaLista.append(salario)

filhos = int(input("Digite quantidade de filhos >>> "))
novaLista.append(filhos)

anosTrabalhados = int(input("Digite quantidade de anos trabalhados >>> "))
novaLista.append(anosTrabalhados)

faculdade = int(input("Digite se possui faculdade sim(1) ou Não(0) >>> "))
novaLista.append(faculdade)

lista2.append(novaLista)
indice = len(lista2) - 1

for i in range(len(lista2)-1):
  dist.append(( (lista2[indice][1] - lista2[i][1] )**2 + (lista2[indice][2] - lista2[i][2] )**2 + (lista2[indice][3] - lista2[i][3])**2 + (lista2[indice][4] - lista2[i][4])**2 )**0.5)

semelhanca = 0

for x in range(len(dist)):
  if dist[semelhanca] > dist[semelhanca]:
    semelhanca = x

print(f"\nA sua similaridade é com  {lista2[semelhanca][0]}!")
print("-- Informação --")
print(f"\t{lista2[indice][0]} - Salário:[{lista2[indice][1]}] - Qtde filhos:[{lista2[indice][2]}] - Trab.[{lista2[indice][3]}] Facul:[{lista2[indice][4]}]")
print(f"\t{lista2[semelhanca][0]} - Salário:[{lista2[semelhanca][1]}] - Qtde filhos:[{lista2[semelhanca][2]}] - Trab.[{lista2[semelhanca][3]}] Facul:[{lista2[semelhanca][4]}]")
```
